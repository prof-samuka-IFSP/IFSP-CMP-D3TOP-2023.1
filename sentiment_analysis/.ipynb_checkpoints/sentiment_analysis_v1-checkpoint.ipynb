# IMDB Movie Ratings Sentiment Analysis

**Sentiment Analysis** is a use case of _Natural Language Processing (NLP)_ and comes under the category of **text classification**. It involves ***classifying a text*** into various **sentiments**, such as _positive_, _negative_, or _neutral_, _happy_ or _sad_, etc.

**Sentiment analysis** is used in various applications such as _social media monitoring_, _customer feedback analysis_, _brand reputation management_, and _market research_. It helps businesses and organizations _to understand customer feedback_, _improve products and services_, and make _better decisions_ based on customer sentiment.

## 3. Text Preprocessing

```python
import neattext.functions as ntx

def text_preprocessing(text_in: str) -> str:
    text = text_in.lower()
    
    text = ntx.fix_contractions(text)
    text = ntx.remove_punctuations(text)
    text = ntx.remove_stopwords(text)
    text = ntx.remove_numbers(text)
    text = ntx.remove_special_characters(text)
    
    return text
```

```python
from tqdm import tqdm
tqdm.pandas()  # it enables some new progress bar functions/methods for pandas
```

```python
# pre-process the training set
df_train['text-pre'] = df_train['text'].progress_apply(lambda text: text_preprocessing(text))
```

```python
df_train
```

```python
# pre-process the testing set
df_test['text-pre'] = df_test['text'].progress_apply(lambda text: text_preprocessing(text))
```

## 4. Pre-trained Sentiment Analysis (VADER)

**VADER (Valence Aware Dictionary and sEntiment Reasoner)** is a _sentiment analysis tool_ that uses a _pre-defined dictionary of words_ to analyze the sentiment of a given text. The dictionary contains _thousands of words and phrases_ that are associated with **positive or negative sentiment**.

When analyzing a _text_ with **VADER**, the tool assigns **a score** to _each word_ based on its **sentiment** in the _dictionary_. The **scores range from -4 to +4**, with negative scores indicating negative sentiment and positive scores indicating positive sentiment.

**VADER** also takes into account the **intensity of the sentiment** conveyed by a word, by assigning a higher score to words that are more strongly associated with positive or negative sentiment.

After assigning scores to each word in the text, VADER _combines them_ to calculate **an overall sentiment score**, known as the ***compound score***. The compound score ranges from -1 (most negative) to +1 (most positive), with a score of 0 indicating neutral sentiment.

**Reference** <br/>
Hutto, C. J., & Gilbert, E. (2014). VADER: A parsimonious rule-based model for sentiment analysis of social media text. In Proceedings of the Eighth International AAAI Conference on Weblogs and Social Media (pp. 216-225).

### 4.1. NLTK Vader
- https://www.nltk.org/_modules/nltk/sentiment/vader.html
- https://www.analyticsvidhya.com/blog/2021/06/vader-for-sentiment-analysis/

```python
import nltk
```

```python
# download vader model
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sent_analyzer = SentimentIntensityAnalyzer()
```

### 4.2. Executing Vader

```python
text = "That is an amazing movie!"
```

```python
sent_analyzer.polarity_scores(text)
```

We have a score for:
- **Negative** sentiment
- **Neutral** sentiment
- **Positive** sentiment
- A **compound** score

The **compound score** in the _VADER sentiment analyzer_ is a _metric_ that represents **the overall sentiment** of a given _text_.

The _score_ is calculated by taking into account _the positive, negative, and neutral words_ in the text, and then **normalizing the score** to be between _-1_ (most negative) and _+1_ (most positive).

A **compound score** of
- **0.0** indicates a _neutral sentiment_
- close to **+1.0** indicate _a more positive sentiment_
- close to **-1.0** indicate _a more negative sentiment_

```python
text = 'I hated that movie. AAAAARG'
sent_analyzer.polarity_scores(text)
```

```python
text = "That's an ok movie"
sent_analyzer.polarity_scores(text)
```

### 4.3. Classifying the Sentiment by Vader

```python
def classify_sentiment(text: str, sent_analyzer) -> int:
    scores_dict = sent_analyzer.polarity_scores(text)
    
    # positive (1) if 'compound' > 0.0
    # negative (0) otherwise
    predicted_sentiment = (scores_dict['compound'] > 0.0)
    
    return predicted_sentiment
```

#### **Train Set**

```python
# classify the training set
df_train['prediction'] = df_train['text-pre'].progress_apply(lambda text: classify_sentiment(text, sent_analyzer))
```

```python
df_train
```

```python
from sklearn.metrics import classification_report

print(classification_report(df_train['label'], df_train['prediction']))
```

#### **Test Set**

```python
# classify the testing set
df_test['prediction'] = df_test['text-pre'].progress_apply(lambda text: classify_sentiment(text, sent_analyzer))
```

```python
from sklearn.metrics import classification_report

print(classification_report(df_test['label'], df_test['prediction']))
```

### 4.4. Pros and Cons of VADER Sentiment Analyzer

**Pros**
- **Easy to use**: VADER is a straightforward and user-friendly tool that requires no training or expertise in natural language processing.
- **Good performance for social media**: VADER was specifically designed to perform well on short, informal texts like social media posts, where traditional sentiment analysis techniques often struggle.
- **Handles emojis and emoticons**: VADER can interpret the sentiment of emojis and emoticons, which are common in social media posts and often carry meaning that is difficult to capture with traditional text-based sentiment analysis.
- **Open source**: VADER is an open-source tool, which means that anyone can use and modify it for their own needs.

**Cons**
- **Limited applicability:** VADER was designed for sentiment analysis of social media posts and may not perform as well on other types of text, such as formal documents or literature.
- **No contextual understanding:** VADER does not have a deep understanding of the context in which words are used, which can lead to inaccuracies in sentiment analysis.
- **Limited language support:** VADER was originally developed for English text and may not perform as well on other languages.
- **Over-reliance on dictionary:** VADER relies heavily on pre-defined sentiment dictionaries, which may not be accurate or up-to-date for all domains or contexts.
